**Classification**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
data=load_breast_cancer()
x=data.data
y=data.target
```

```python
x=pd.DataFrame(x,columns=data.feature_names)
y=pd.Series(y)
```

```python
print(type(x))
print(type(y))
```

```python
x.isna().sum()
```

```python
y.isna().sum()
```

```python
x.head()
```

**No null values present is this dataset**

```python
model_accuracy={}
```

**Created a dictionary to store the accuracy score of each model**

```python
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

**Using StandardScaler for scaling the values in the data and make it equal base to a certain value** 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
l_model=LogisticRegression()
l_model
```

```python
l_model.fit(x_train,y_train)
l_model
```

```python
l_y_pred=l_model.predict(x_test)
```

```python
l_accuracyscore=accuracy_score(y_test,l_y_pred)
print(l_accuracyscore)
```

```python
model_accuracy["Logistic Regression"]=l_accuracyscore
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,l_y_pred))
```

```python
print("Classification Report \n",classification_report(y_test,l_y_pred))
```

**Logistic regression is a classification algorithm that models the probability of a class by applying a sigmoid function to a linear combination of input features.**

**During training, it learns the optimal weights for each feature so that the predicted probabilities match the actual labels as closely as possible. It does this by fitting a linear equation to the data and then passing the result through a sigmoid function, which converts it into values between 0 and 1.**

**During testing, the model uses the learned parameters (weights) to compute the probability for new data points. Based on a threshold (commonly 0.5), it classifies the output into one of the target classes.**

```python
dt=DecisionTreeClassifier(
    criterion='entropy',
    max_depth=4,
random_state=42)
dt
```

```python
dt.fit(x_train,y_train)
dt
```

```python
dt_y_pred=dt.predict(x_test)
```

```python
dt_accuracyscore=accuracy_score(y_test,dt_y_pred)
print(dt_accuracyscore)
```

```python
model_accuracy["Decision Tree Classifier"]=dt_accuracyscore
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,dt_y_pred))
```

```python
print("Classification Report \n",classification_report(y_test,dt_y_pred))
```

**A Decision Tree Classifier builds a tree-like structure by repeatedly splitting the dataset based on feature values. At each node, it asks a question about a feature (e.g., "Is feature X ≤ value?") to divide the data into subsets.**

**The model selects the best feature for splitting by using measures like entropy or Gini impurity. These measures help determine how well a feature separates the data into different classes.**
  


```python
Rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
Rf_model
```

```python
Rf_y_pred=Rf_model.predict(x_test)
```

```python
model_accuracy["Random Forest Classifier"]=Rf_accuracyscore
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,Rf_y_pred))
```

```python
print("Classification Report \n",classification_report(y_test,Rf_y_pred))
```

**K-Nearest Neighbors (KNN) classifier classifies a data point based on the majority class among its k nearest neighbors in the feature space. It uses distance metrics (like Euclidean distance) to find the closest points.**

```python
svm=SVC(kernel='rbf',C=1.0,gamma='scale')
svm
```
